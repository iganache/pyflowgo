```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
vfile = "~/Projects/pyflowgo/results_flowgo/results_main_flowgo_Tolbachik_Flynn_Venus_700m3s.csv"
efile = "~/Projects/pyflowgo/results_flowgo/results_main_flowgo_Tolbachik_Flynn_Earth_700m3s.csv"
```

```python
vdf = pd.read_csv(vfile)
vdf.columns
vdf['position'] = vdf['position']/1e3

edf = pd.read_csv(efile)
edf.columns
edf['position'] = edf['position']/1e3
```

```python
vdf.columns
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(9,4)
vdf.plot('position', 'viscosity', ax=ax, label = "Venus")
edf.plot('position', 'viscosity', ax=ax, label = "Earth")
plt.semilogy()
ax.set_ylabel("viscosity (Pas)")
ax.set_xlabel("distance (km)")
plt.savefig("viscosity.png")
plt.show()
```

```python
fig, ax = plt.subplots(2,1)
fig.set_size_inches(9,9)
vdf.plot('position', 'flowgofluxsnyderheat', ax=ax[0], label = "Venus")
edf.plot('position', 'flowgofluxradiationheat', ax=ax[0], label = "Earth")
ax[0].semilogy()

vdf.plot('position', 'flowgofluxsnyderheat', ax=ax[1], label = "Venus")
edf.plot('position', 'flowgofluxforcedconvectionheat', ax=ax[1], label = "Earth")
ax[1].semilogy()
plt.show()
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(9,4)
vdf.plot('position', 'core_temperature', ax=ax, label = "Venus")
edf.plot('position', 'core_temperature', ax=ax, label = "Earth")
# plt.semilogy()
ax.set_ylabel("core_temperature (K)")
ax.set_xlabel("distance (km)")
plt.savefig("coretemp.png")
plt.show()
```
